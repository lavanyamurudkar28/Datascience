```python
import matplotlib.pyplot as plt
#frequencies
ages = [2,570,40,30,45,50,45,43,40,44,60,7,13,57,18,90,77,32,21,40]

#setting the ranges and no. of intervals
range = (0,100)
bins  =  10

#plotting a histogram
plt.hist(ages, bins, range,
color = "blue", histtype="bar", edgecolor = "black")
plt.xlabel('age')

#frequency label
plt.ylabel('no.of people')

#plot title
plt.title('my histogram')
```

```python
import matplotlib.pyplot as plt

#defining labels
activities = ['eat','sleep','work','play']

#portion covered by eah label
slices = [3,7.8,6]

#color for eah label
colors = ['r', 'y', 'g','b']

#plotting the pie chart
plt.pie(slices, labels = activities, colors=colors, startangle=30, shadow = True, explode = (0.1,0,0,0), radius=1.2, autopct = '%1.1f%%')

plt.legend()
plt.show()
```
